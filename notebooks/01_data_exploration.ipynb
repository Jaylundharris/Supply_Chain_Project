```python
conn.execute("DROP VIEW IF EXISTS orders_clean;")
conn.execute("DROP VIEW IF EXISTS supplier_clean;")
conn.execute("DROP VIEW IF EXISTS inventory_clean;")
conn.execute("DROP VIEW IF EXISTS warehouse_clean;")
```

```python
orders_clean_sql = """
CREATE VIEW IF NOT EXISTS orders_clean AS
SELECT
    order_id,
    sku_id,
    warehouse_id,
    CAST(order_timestamp AS DATE) AS order_date,
    CAST(fulfillment_timestamp AS DATE) AS fulfillment_date,
    region,
    shipping_method,
    quantity,
    fulfillment_time_hours,
    delivery_time_days,
    -- NEW column: time from order to fulfillment
    (JULIANDAY(fulfillment_timestamp) - JULIANDAY(order_timestamp)) AS days_to_fulfill
FROM orders;
"""
conn.execute(orders_clean_sql)

pd.read_sql("SELECT * FROM orders_clean LIMIT 5", conn)
```

```python
supplier_clean_sql = """
CREATE VIEW supplier_clean AS
SELECT
    supplier_id,
    shipment_id,
    CAST(order_date AS DATE)        AS order_date,
    promised_lead_time,
    actual_lead_time,
    units_ordered,
    units_received,
    delivery_status,
    cost_per_unit,
    -- NEW: safe fill rate
    CASE 
        WHEN units_ordered = 0 THEN NULL
        ELSE units_received * 1.0 / units_ordered
    END AS fill_rate,
    -- NEW: lead time variance (actual - promised)
    actual_lead_time - promised_lead_time AS lead_variance
FROM suppliers;
"""
conn.execute(supplier_clean_sql)

pd.read_sql("SELECT * FROM supplier_clean LIMIT 5;", conn)

```

```python
inventory_clean_sql = """
CREATE VIEW inventory_clean AS
SELECT
    movement_id,
    sku_id,
    warehouse_id,
    movement_type,
    quantity,
    CAST(timestamp AS DATE)         AS movement_date,
    starting_inventory_level,
    ending_inventory_level,
    reorder_point,
    safety_stock,
    -- NEW: change in inventory level
    ending_inventory_level - starting_inventory_level AS inventory_change
FROM inventory;
"""
conn.execute(inventory_clean_sql)

pd.read_sql("SELECT * FROM inventory_clean LIMIT 5;", conn)
```

```python
sku_clean_sql = """
CREATE VIEW IF NOT EXISTS sku_clean AS
SELECT
    sku_id,
    sku_name,
    category,
    sub_category,
    cost,
    price,
    supplier_id,
    weight,
    price - cost AS profit_per_unit,
    CASE 
        WHEN cost = 0 THEN NULL
        ELSE (price - cost) * 1.0 / cost
    END AS margin_pct
FROM sku_master;
"""
conn.execute(sku_clean_sql)
pd.read_sql("SELECT * FROM sku_clean LIMIT 5", conn)
```

```python
warehouse_clean_sql = """
CREATE VIEW warehouse_clean AS
SELECT
    warehouse_id,
    CAST(date AS DATE) AS work_date,
    picks_per_hour,
    lines_per_hour,
    receiving_volume,
    outbound_volume,
    labor_hours,
    errors,
    rework,
    -- NEW: productivity (guard against zero hours)
    CASE 
        WHEN labor_hours = 0 THEN NULL
        ELSE outbound_volume * 1.0 / labor_hours
    END AS units_per_labor_hour,
    -- NEW: error rate (guard against zero outbound volume)
    CASE 
        WHEN outbound_volume = 0 THEN NULL
        ELSE errors * 1.0 / outbound_volume
    END AS error_rate
FROM warehouse;
"""
conn.execute(warehouse_clean_sql)

pd.read_sql("SELECT * FROM warehouse_clean LIMIT 5;", conn)
```

```python
pd.read_sql("""
    SELECT name, type 
    FROM sqlite_master 
    WHERE type IN ('table', 'view')
    ORDER BY type, name;
""", conn)
```

```python
pd.read_sql("SELECT * FROM orders_clean", conn).to_csv("../data/processed/orders_clean.csv", index=False)
pd.read_sql("SELECT * FROM supplier_clean", conn).to_csv("../data/processed/supplier_clean.csv", index=False)
pd.read_sql("SELECT * FROM inventory_clean", conn).to_csv("../data/processed/inventory_clean.csv", index=False)
pd.read_sql("SELECT * FROM warehouse_clean", conn).to_csv("../data/processed/warehouse_clean.csv", index=False)
pd.read_sql("SELECT * FROM sku_clean", conn).to_csv("../data/processed/sku_clean.csv", index=False)
```
